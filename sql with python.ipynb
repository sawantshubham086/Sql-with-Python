```python
import mysql.connector
```

```python
x = mysql.connector.connect(host='localhost',user='root',password='',database='world')
```

```python
import pandas as pd
```

```python
pd.read_sql_query("select * from city",x)
```

```python
# Order by clause
pd.read_sql_query("select Name,population from city order by population asc",x)
```

```python
#column expression with arithmatical operator
pd.read_sql_query("select Name,population-10000 from city where countrycode='AFG'",x)
```

```python
#Logical operators
pd.read_sql_query("select District from city where population < 100000 and countrycode in ('AFG','NLD')",x)
```

```python
# Grop by function and average function
pd.read_sql_query("select District,count(CountryCode) from city where population <100000 group by district",x)
```

```python
pd.read_sql_query("SELECT * FROM country",x)
```

```python
pd.read_sql_query("SELECT * FROM countrylanguage",x)
```

```python
# Set operators
pd.read_sql_query("SELECT CountryCode FROM countrylanguage intersect SELECT CountryCode FROM city",x)
```

```python
#Join
pd.read_sql_query("SELECT cl.Language,c.Name FROM countrylanguage cl join city c on (c.CountryCode=c.CountryCode)",x)
```

```python
#Subquerry
pd.read_sql_query("SELECT * FROM countrylanguage where CountryCode= any (select CountryCode from city where CountryCode='AFG')" ,x)
```

```python
# Having Clause
pd.read_sql_query("SELECT CountryCode,count(Language) FROM countrylanguage group by CountryCode having countrycode in ('AFG','ABW')",x)
```
